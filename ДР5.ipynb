```python
from sklearn.preprocessing import LabelEncoder

# Создание экземпляра LabelEncoder
label_encoder = LabelEncoder()

# Замена категориальных признаков на числовые
df['Company\nLocation'] = label_encoder.fit_transform(df['Company\nLocation'])
df['Bean\nType'] = label_encoder.fit_transform(df['Bean\nType'])
df['Broad Bean\nOrigin'] = label_encoder.fit_transform(df['Broad Bean\nOrigin'])

# Проверка изменений
print(df.head())

```

```python
df['Company \n(Maker-if known)'] = label_encoder.fit_transform(df['Company \n(Maker-if known)'])
df['Specific Bean Origin\nor Bar Name'] = label_encoder.fit_transform(df['Specific Bean Origin\nor Bar Name'])
```

```python

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.histplot(df['Rating'], bins=10, kde=True)
plt.title('Распределение рейтингов шоколада')
plt.xlabel('Рейтинг')
plt.ylabel('Частота')
plt.show()


plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Cocoa\nPercent', y='Rating')
plt.title('Распределение рейтингов по проценту какао')
plt.xlabel('Процент какао')
plt.ylabel('Рейтинг')
plt.show()


plt.figure(figsize=(12, 8))
sns.boxplot(data=df, x='Company\nLocation', y='Rating')
plt.title('Распределение рейтингов по странам')
plt.xlabel('Страна')
plt.ylabel('Рейтинг')
plt.xticks(rotation=90)
plt.show()

# Корреляционная матрица
corr = df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Корреляционная матрица')
plt.show()
```
